```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set visualization style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12, 6)

# Load datasets
client_df = pd.read_csv("client_data (1) (1).csv")  
price_df = pd.read_csv("price_data (1) (1).csv")    

# --- Step 1: Basic Info ---
print("Client Data Info:")
print(client_df.info())
print("\nPrice Data Info:")
print(price_df.info())

# --- Step 2: Descriptive Statistics ---
print("\nClient Data - Descriptive Statistics:")
print(client_df.describe())

# --- Step 3: Distribution Plots for Key Numeric Columns ---
numeric_cols = [
    "cons_12m", "cons_gas_12m", "cons_last_month", "forecast_cons_12m",
    "forecast_price_energy_off_peak", "forecast_price_energy_peak",
    "margin_net_pow_ele", "net_margin", "pow_max"
]

# Create histograms with KDE for numeric columns
fig, axes = plt.subplots(nrows=3, ncols=3, figsize=(20, 15))
axes = axes.flatten()

for i, col in enumerate(numeric_cols):
    sns.histplot(client_df[col], bins=30, kde=True, ax=axes[i])
    axes[i].set_title(f'Distribution of {col}')
    axes[i].set_xlabel(col)

plt.tight_layout()
plt.show()

# --- Step 4: Churn Rate Summary ---
churn_rate = client_df["churn"].value_counts(normalize=True)
print("\nChurn Rate:")
print(churn_rate)

# Optional: Visualize churn distribution
sns.countplot(data=client_df, x="churn")
plt.title("Churn Count")
plt.xlabel("Churn (0 = No, 1 = Yes)")
plt.ylabel("Number of Customers")
plt.show()
```
